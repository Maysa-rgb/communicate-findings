## Diamond dataset contains some quality standards variables like cut , clarity and color. in addition to weight of diamond pieces in carat ,price and some more information like dimensions of each piece .

## This presentation will show three derived insights from diamond data :
- **Cut quality standard is the most controllable quality scale.**
- **The relationship between cut and price isn't absolute but other factors play role.**
- **The relationship between cut , price and weight.**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
%matplotlib inline
```

```python
diamond = pd.read_csv('diamonds.csv')
```

```python
order = {'cut':['Fair','Good','Very Good','Premium','Ideal'] ,
        'color':['J', 'I', 'H', 'G', 'F', 'E', 'D'],
        'clarity' :['I1', 'SI2', 'SI1', 'VS2', 'VS1', 'VVS2', 'VVS1', 'IF']}
for var in order:
    ordered = pd.api.types.CategoricalDtype(ordered = True , categories = order[var])
    diamond[var] = diamond[var].astype(ordered)
```

# First visualization :

- **Cut quality standard is the most controllable quality scale.**

```python
plt.figure(figsize=[20,6])

#plot 1
plt.subplot(1,3,1) # 1 row , 3 columns , plot 1
base_color = sb.color_palette()[0]
sb.countplot(data= diamond , x='cut', color= base_color)
plt.title('production of cut quality levels',fontsize = 18)
plt.xlabel('Cut levels',fontsize = 16)
plt.xticks(fontsize = 12)
plt.yticks(fontsize = 12);

# plot 2
plt.subplot(1,3,2)
sb.countplot(data= diamond , x='clarity', color= base_color)
plt.title('Production of clarity quality levels', fontsize = 18)
plt.xlabel('Clarity levels',fontsize =16)
plt.xticks(fontsize = 12)
plt.yticks(fontsize = 12);

# plot 3
plt.subplot(1,3,3)
sb.countplot(data= diamond , x='color', color= base_color)
plt.title('Production of color quality levels', fontsize = 18)
plt.xlabel('Color levels',fontsize = 16)
plt.xticks(fontsize = 12)
plt.yticks(fontsize = 12);

```

# Comment
**These three plots show the production of diamond over three different quality categories.each category has quality levels orderd from the least on the left to the most quality on the right.**
- **the production of the highest quality level was met only in the cut quality production where  ideal cut occupied the first  production level.**

- **on the other hand we find in clarity scale we find IF (the best quality) production occupy the seventh level of production.**

- **finally in color scale we find D (best color level) occupy the fifth production level.**

**I believe that brightness of the diamond piece which is affected by how it's cut has strong impact on diamond market , in addition cut quality is the most controllable quality standard as it's subject to human factor , while clarity and color subject to the availability of diamond in nature and mining process , hence ideal cut quality occupied the first production among the other quality standards.**


# Second visualization :
- **The relationship between cut and price isn't absolute but other factors play role.**

```python
base_color = sb.color_palette()[0]
plt.figure(figsize = [15,8])
sb.boxplot(data = diamond , y='price', x='cut' ,showfliers = False , color = base_color)
plt.ylabel('price $',fontsize = 13)
plt.xlabel('cut levels',fontsize = 13)
plt.title('relationship between cut quality and price',fontsize = 15);
```

# Comment

**We could see unexpectedly ideal diamond didn't occupy the highest price statistics as shown in the boxplot and how lower quality levels occupied higher prices which means the relationship between the quality of cut and price isn't absolute and some other factors plays role like weight as shown in the next plot.** 

# Third visualization:

- **The relationship between cut , price and weight.** 

```python
g = sb.FacetGrid(data = diamond, hue = 'cut', palette = 'colorblind',height = 8)
g.map(plt.scatter, 'carat', 'price')
g.add_legend(fontsize = 13)
plt.ylabel('price $', fontsize = 15)
plt.xlabel('weight (carat)',fontsize = 13)
plt.title('cut quality effect on price and weight relationship',fontsize = 15);
```

# Comment

**The scatter plot shows how weight affects the relationship between cut quality and price ,and also cut quality affects the relationship between price and weight meaning it's mutual effect.**


# Summary 

- Cut quality standard is the most controllable and hence it's highest quality level is the most produced.

- There is no direct absolute correlation between cut quality standard and diamond price .

- Diamond price doesn't depend on one factor but it's determined based on many factors comes among them price ,quality standards and weight as was discussed.
